```python
import pandas as pd
```

```python
data = [['Kashvi',38,40,40,38],
        ['Riya',18,30,29,26],
        ['Nandini',38,39,20,32],
        ['Daksh',39,40,40,38],
        ['Kathan',15,36,20,25]]

kd=pd.DataFrame(data, columns = ['STUDENTNAME','Maths','English','Science','DATASCIENCE'])
kd
```

```python
fig, axes = plt.subplots(2,2,layout="constrained")
axes[0,0].plot(kd.STUDENTNAME,kd.Maths,'m*-',label='Maths')
axes[0,0].set_xlabel('Name')
axes[0,0].set_ylabel('Maths')
axes[0,0].set_title("Maths")


axes[0,1].plot(kd.STUDENTNAME,kd.English,'b-',label='Eng')
axes[0,1].set_xlabel('Name')
axes[0,1].set_ylabel('ENGLISH')
axes[0,1].set_title("ENGLISH")


axes[1,0].plot(kd.STUDENTNAME,kd.Science,'y*-',label='SCI')
axes[1,0].set_xlabel('Name')
axes[1,0].set_ylabel('SCIENCE')
axes[1,0].set_title("SCIENCE")


axes[1,1].plot(kd.STUDENTNAME,kd.DATASCIENCE,'gv-',label='DSC')
axes[1,1].set_xlabel('Name')
axes[1,1].set_ylabel('DATA SCIENCE')
axes[1,1].set_title("DATA SCIENCE")




```

```python
np.random.seed(3)
x = 4 + np.random.normal(0,2,24)
y = 4 + np.random.normal(0,2,len(x))
 
sizes = np.random.uniform(45,80, len(x))
colors = np.random.uniform(15,80,len(x))

fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c = colors)
ax.set(xlim=(0,8), ylim=(0,8))
```
